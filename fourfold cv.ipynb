```python
def run_random_forest(iterations):
    total_scores = pd.DataFrame(columns = ['set', 'MSE', 'MAE', 'r^2'])
    for j in range(iterations):
        X_subset1, X_subset2, Y_subset1, Y_subset2 = train_test_split(X, Y_is, test_size = 0.5, random_state = random_state)

        X_1, X_2, Y_1, Y_2 = train_test_split(X_subset1, Y_subset1, test_size = 0.5, random_state = random_state)
        X_3, X_4, Y_3, Y_4 = train_test_split(X_subset2, Y_subset2, test_size = 0.5, random_state = random_state)

        Xs = [X_1, X_2, X_3, X_4]
        Ys = [Y_1, Y_2, Y_3, Y_4]

        scores = pd.DataFrame(columns = ['set', 'MSE', 'MAE', 'r^2'])
        for i in range(4): 

            forest = RandomForestRegressor()

            X_test = Xs[i]
            Y_test = Ys[i]
            X_train = X.drop(list(X_1.index))
            Y_train = Y_is.drop(list(X_1.index))

            forest.fit(X_train.reset_index().drop(['Reaction'], axis = 1), Y_train.reset_index().drop('Reaction', axis = 1).values.ravel())
            Y_pred = forest.predict(X_test.reset_index().drop(['Reaction'], axis = 1))

            results = pd.concat([Y_test.reset_index(), pd.DataFrame(Y_pred)], axis = 1)
            results.set_index('Reaction', inplace = True)
            results.columns = ['Actual', 'Predicted']


            scores = scores.append({'set': str(i+1),
                        'MSE': mean_squared_error(Y_test, Y_pred),
                        'MAE': mean_absolute_error(Y_test, Y_pred),
                        'r^2': r2_score(Y_test, Y_pred)}, ignore_index = True)


        averages = pd.DataFrame(scores.mean(axis = 0)).T
        scores = scores.append({'set': 'average',
                                'MSE': averages.iloc[0,1],
                                'MAE': averages.iloc[0,2],
                                'r^2': averages.iloc[0,3]}, ignore_index = True)
        scores.set_index('set', inplace = True)
        total_scores = total_scores.append({'set': str(int(j+1)),
                             'MSE': scores.loc['average', 'MSE'],
                             'MAE': scores.loc['average', 'MAE'],
                             'r^2': scores.loc['average', 'r^2']}, ignore_index = True)
        
    total_averages = pd.DataFrame(total_scores.mean(axis = 0)).T
    total_scores = total_scores.append({'set': 'average',
                                'MSE': total_averages.iloc[0,1],
                                'MAE': total_averages.iloc[0,2],
                                'r^2': total_averages.iloc[0,3]}, ignore_index = True)
        
    total_scores.set_index('set', inplace = True)

    return total_scores
```

```python
rf_score = run_random_forest(100)
```
