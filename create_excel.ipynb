```python
def add_to_dataframe(dataframe, folder, foldername, category = None):
    reaction = Reaction(folder)
    length = reaction.length()
    for i in range(1, length + 1):
        dataframe = dataframe.append({
        'reaction': foldername,
        'category': category,
        'entry': i,
        'substrate1': reaction.substrate1()[i],
        'substrate2': reaction.substrate2()[i],
        'solvent': reaction.solvent()[i],
        'product': reaction.product()[i],
        'catalyst': reaction.catalyst()[i]
        }, ignore_index = True)
    return dataframe
```
